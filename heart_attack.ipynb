```python
from scipy.stats import ttest_ind

# T-test for heart attack risk by smoking status
smokers = cleaned_wlb_heart[cleaned_wlb_heart['smoking'] == 1]['heart_attack_risk']
non_smokers = cleaned_wlb_heart[cleaned_wlb_heart['smoking'] == 0]['heart_attack_risk']
t_stat, p_value = ttest_ind(smokers, non_smokers)

print(f'T-test results: t-statistic = {t_stat}, p-value = {p_value}')
```
